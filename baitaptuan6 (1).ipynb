```python
#W6A9
'''Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key.'''
input1 = input().split()
dict1 = {item.split(':')[0]: int(item.split(':')[1]) for item in input1}

input2 = input().split()
dict2 = {item.split(':')[0]: int(item.split(':')[1]) for item in input2}

combined_dict = dict1.copy()
for key, value in dict2.items():
    if key in combined_dict:
        combined_dict[key] += value
    else:
        combined_dict[key] = value
sorted_keys = sorted(combined_dict.items())
result_dict = dict(sorted_keys)
print(result_dict)
```

```python
#W6A10
def find_pairs_with_sum(a, k):
    pairs = []

    n = len(a)
    for i in range(n):
        for j in range(i + 1, n):
            if a[i] + a[j] == k:
                pair = (a[i], a[j])
                if pair not in pairs:
                    pairs.append(pair)

    pairs.sort(key=lambda x: x[0])
    return pairs

a = list(map(int, input().split()))
k = int(input("k: "))
result = find_pairs_with_sum(a, k)
print(result)

```

```python
#W6A14
a = sorted(list(map(int, input().split())))
b = sorted(list(map(int, input().split())))
giao = []
i, j = 0, 0
while i < len(a) and j < len(b):
    if a[i] <= b[j]:
        giao.append(a[i])
        i += 1
    else:
        giao.append(b[j])
        j += 1
while i < len(a):
    giao.append(a[i])
    i += 1
while j < len(b):
    giao.append(b[j])
    j += 1
print(giao)
    
    
```
