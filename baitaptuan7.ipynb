```python
#W7A7
def sub(a):
    n = len(a)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if a[j] < a[i]:
                dp[i] = max(dp[i], dp[j] +1)
    return max(dp)
```

```python
#W7A8
import heapq

def minInterval(intervals, queries):
    intervals.sort(key=lambda x: x[0])
    sorted_queries = sorted([(q, i) for i, q in enumerate(queries)])
    res = [-1] * len(queries)
    heap = []  
    i = 0
    for q, idx in sorted_queries:
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            length = r - l + 1
            heapq.heappush(heap, (length, r))
            i += 1
        while heap and heap[0][1] < q:
            heapq.heappop(heap)
        if heap:
            res[idx] = heap[0][0]
        else:
            res[idx] = -1
    return res
intervals = [[1,4],[2,4],[3,6],[4,4]]
queries = [2,3,4,5]

print(minInterval(intervals, queries))
```

```python
#W7A9
'''Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải
nhỏ hơn nums[i]. Trả về danh sách các số đếm. '''
def binary_search(arr, x):
    l, r = 0, len(arr)
    while l < r:
        mid = (l + r) // 2
        if arr[mid] < x:
            l = mid + 1
        else:
            r = mid
    return l

def countSmaller(nums):
    n = len(nums)
    res = [0] * n
    sorted_list = []

    for i in range(n - 1, -1, -1):
        pos = binary_search(sorted_list, nums[i])
        res[i] = pos
        sorted_list.insert(pos, nums[i])

    return res
s = input().strip()

nums = []
num = ""
for c in s:
    if c.isdigit() or c == '-':
        num += c
    else:
        if num != "":
            nums.append(int(num))
            num = ""
if num != "":
    nums.append(int(num))

print(countSmaller(s))
```
