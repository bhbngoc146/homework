```python
#W7A16
def mangdep(a):
    freq = {}
    s = 0
    for x in a:
        freq[x] = freq.get(x,0) + 1
    for key, value in freq.items():
        if key < value:
            s += value - key
        elif key>value:
             s += value
    if s == 0:
        return True
    else:
        return 'No', abs(s)
a = list(map(int, input().split()))
if mangdep(a) == True:
    print('Yes')
else:
    print(*mangdep(a))
```

```python
#W7A17
def majority_element(a):
    candidate = None
    count = 0

    for x in a:
        if count == 0:
            candidate = x
            count = 1
        elif x == candidate:
            count += 1
        else:
            count -= 1
    if candidate is not None and a.count(candidate) > len(a) // 2:
        print(candidate)


a = list(map(int, input().split()))

majority_element(a)
```

```python
#7A21
n1 = [1,3,5,7,8,10,12]
n2 = [4,6,9,11]
def tinhngay(T):
    s1, s2, s3, ss = 0, 0, 0, 0
    for i in range(1, T):
        if i in n1:
            s1 += 31
        elif i in n2:
            s2 += 30
        elif i == 2:
            s3 += 29
    ss = s1 + s2 + s3
    return ss
N, T = map(int, input().split())
if T in n1 and N <= 31:
    print(tinhngay(T)+N)
elif T in n2 and N <= 30:
    print(tinhngay(T)+N)
elif T == 2 and N <= 28:
    print(tinhngay(T)+N)
else:
    print('Invalid')
```

```python
#W7A27
s = input().strip()
st = []
match = {')': '(', ']': '[', '}': '{'}
ok = True
for ch in s:
    if ch in "([{":
        st.append(ch)
    elif ch in ")]}":
        if not st or st[-1] != match[ch]:
            ok = False
            break
        st.pop()
if ok and not st:
    print("YES")
else:
    print("NO")
```

```python
#W7A28
def longest_composite_sequence(N):
    if N < 4:
        return []  

    is_prime = [True] * (N + 1)
    is_prime[0] = is_prime[1] = False

    for i in range(2, int(N ** 0.5) + 1):
        if is_prime[i]:
            for j in range(i * i, N + 1, i):
                is_prime[j] = False
    max_len = 0
    max_start = -1
    current_len = 0
    current_start = 0
    for i in range(1, N + 1):
        if i > 1 and not is_prime[i]: 
            if current_len == 0:
                current_start = i
            current_len += 1
        else:
            if current_len > max_len:
                max_len = current_len
                max_start = current_start
            current_len = 0
    if current_len > max_len:
        max_len = current_len
        max_start = current_start

    return list(range(max_start, max_start + max_len))
N = 10
print(longest_composite_sequence(N))
```

```python
#W7A29
# Nhập danh sách
a = list(map(int, input().split()))

n = len(a)

max_len = 1
max_start = 0
cur_len = 1
cur_start = 0
for i in range(1, n):
    if a[i] > a[i - 1]:
        cur_len += 1
    else:
        if cur_len > max_len:
            max_len = cur_len
            max_start = cur_start
        cur_len = 1
        cur_start = i
if cur_len > max_len:
    max_len = cur_len
    max_start = cur_start

print(*a[max_start:max_start + max_len])
```

```python
#W7A29
a = list(map(int, input().split()))
n = len(a)

if n == 0:
    print([])
else:
    dp = [[x] for x in a] 

    for i in range(n):
        for j in range(i):
            if a[j] < a[i] and len(dp[j]) + 1 > len(dp[i]):
                dp[i] = dp[j] + [a[i]]
    longest = dp[0]
    for seq in dp:
        if len(seq) > len(longest):
            longest = seq
    print(*longest)
```
