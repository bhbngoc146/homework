```python
#W4A2
import math

def snt(a):
    if a < 2:
        return False
    for i in range(2, int(math.sqrt(a)) + 1):
        if a % i == 0:
            return False
    return True

n = -1
while n <= 0:
    n = int(input("Nhập số nguyên dương: "))

if snt(n):
    print(n, "là số nguyên tố")
else:
    print(n, "không phải số nguyên tố")
```

```python
#W4A8
def reverse_number(n):
    return int(str(n)[::-1])

def is_palindrome(n):
    return str(n) == str(n)[::-1]

n = int(input("Nhập số nguyên dương n: "))

steps = 0

while not is_palindrome(n):
    n = n + reverse_number(n)
    steps += 1

print("Số bước:", steps)
print("Giá trị palindrome:", n)
```

```python
#W4A9
n = int(input("Nhập n: "))

result = []

k = 1
while k * k <= n:
    sq = k * k
    s = str(sq)
    if len(s) == len(set(s)):  
        result.append(sq)
    k += 1

if len(result) == 0:
    print("(no number)")
else:
    print(*result)
```

```python
#W4A10
def collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        length += 1
    return length

n = int(input("Nhập n: "))

best_x = 1
max_len = collatz_length(1)

for x in range(2, n + 1):
    L = collatz_length(x)
    if L > max_len:
        max_len = L
        best_x = x
print(best_x, max_len)
```

```python
#W4A13
try:
    a = int(input())
    b = int(input())

    tong_uoc_a = 0
    for i in range(1, a//2 + 1):      
        if a % i == 0:
            tong_uoc_a += i

    tong_uoc_b = 0
    for i in range(1, b//2 + 1):      
        if b % i == 0:
            tong_uoc_b += i

    if tong_uoc_a == b and tong_uoc_b == a:
        print("True")
    else:
        print("False")

except ValueError:
    print("Bạn nhập sai dữ liệu")
```
