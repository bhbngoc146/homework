```python
# W5A1
def Max_Of_Two(a: int, b: int) -> int:
    return a if a > b else b

print(Max_Of_Two(3, 5))
```

```python
# W5A2
from typing import Tuple

def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a

print(swap(2, 7))
```

```python
# W5A3
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(17))
```

```python
# W5A4
def is_perfect(n: int) -> bool:
    if n <= 0:
        return False
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n

print(is_perfect(28))
```

```python
# W5A5
def find_first(lst, k: int) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

print(find_first([1, 3, 5, 3], 3))
```

```python
# W5A6
def factorial(n: int) -> int:
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```

```python
# W5A7
def calculator(num1: float, num2: float, operat: str) -> float:
    if operat == '+':
        return round(num1 + num2, 2)
    elif operat == '-':
        return round(num1 - num2, 2)
    elif operat == '*':
        return round(num1 * num2, 2)
    elif operat == '/':
        if num2 == 0:
            raise ValueError("Không thể chia cho 0")
        return round(num1 / num2, 2)
    else:
        raise ValueError("Toán tử không hợp lệ")

print(calculator(5, 3, '/'))
```

```python
# W5A8
def hamming_distance(x: int, y: int) -> int:
    return bin(x ^ y).count('1')

print(hamming_distance(1, 4))
```

```python
# W5A9
def sum_digits(n: int) -> int:
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s

print(sum_digits(12345))
```

```python
# W5A10
def is_isomorphic(a: str, b: str) -> bool:
    map_a, map_b = {}, {}
    for ca, cb in zip(a, b):
        if (ca in map_a and map_a[ca] != cb) or (cb in map_b and map_b[cb] != ca):
            return False
        map_a[ca] = cb
        map_b[cb] = ca
    return True

print(is_isomorphic("abca", "zbxz"))
```
