```python
#W8A1
class Cy:
    def __init__(self,r,h):
        self.r = r
        self.h = h
        self.pi = 3.14
    def sa(self):
        return 2 * self.pi * self.r * self.r + 2 * self.pi * self.r * self.h
    def v(self):
        return self.pi * self.r * self.r * self.h
r, h = map(int, input().split())
cy = Cy(r,h)

S = cy.sa()
V = cy.v()
print(f"{S:.2f} {V:.2f}")
```

```python
#W8A2
class Date:
    def __init__(self,d,m,y):
        self.d = d
        self.m = m
        self.y = y
    def is_leap_year(self):
        return self.y % 400 == 0 or (self.y % 4 == 0 and self.y % 100 != 0)
    def days_in_month(self):
        if self.m == 2:
            return 29 if self.is_leap_year() else 28
        if self.m in [2,4,6,9,11]:
            return 30
        return 31
    def is_valid(self):
        if self.y < 1 or self.m < 1 or self.m > 12:
            return False
        if self.d < 1 or self.d > self.days_in_month():
            return False
        return True
    def next_date(self):
        self.d += 1
        if self.d > self.days_in_month():
            self.d = 1
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
    def format(self):
        return f"{self.d:02d}/{self.m:02d}/{self.y:04d}"
s = input().strip()
d, m, y = map(int, s.split("/"))

date = Date(d, m, y)

if not date.is_valid():
    print("INVALID")
else:
    date.next_date()
    print(date.format())
```

```python
#W8A3
class toado:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def vitri(self):
        if self.x == 0 and self.y == 0:
            return 'goc toa do'
        elif self.x > 0 and self.y == 0:
            return 'Nam tren truc hoanh'
        elif self.x == 0 and self.y > 0:
            return 'Nam tren truc tung'
    def kc(self):
        return (self.x**2+self.y**2)**0.5
x, y = map(int, input().split())
p = toado(x, y)
VT = p.vitri()
KC = p.kc()
print(VT)
print(f"{KC:.2f}")
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h,k):
        self.w = w
        self.h = h
        self.k = k
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self):
        self.w = self.k * self.w 
        self.h = self.k * self.h
w,h,k = map(int, input().split())
rec = Rectangle(w,h,k)
rec.scale()
S = rec.area()
P = rec.perimeter()
print(S,P)
```

```python
#W8A7
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mau so khong duoc bang 0")
        if den < 0:
            num = -num
            den = -den
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g
    def add(self, other):
        return Fraction(
            self.num * other.den + other.num * self.den,
            self.den * other.den
        )
    def sub(self, other):
        return Fraction(
            self.num * other.den - other.num * self.den,
            self.den * other.den
        )
    def mul(self, other):
        return Fraction(
            self.num * other.num,
            self.den * other.den
        )
    def div(self, other):
        return Fraction(
            self.num * other.den,
            self.den * other.num
        )
    def __str__(self):
        return f"{self.num}/{self.den}"


a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)

f1 = Fraction(a, b)
f2 = Fraction(c, d)
if op == '+':
    res = f1.add(f2)
elif op == '-':
    res = f1.sub(f2)
elif op == '*':
    res = f1.mul(f2)
elif op == '/':
    res = f1.div(f2)
print(res)
```

```python
#W8A6
class Shopping:
    def __init__(self):
        self.items = []
    def add(self, name, price):
        self.items.append({'name': name, 'price': price})
    def remove(self,name):
        for item in self.items:
            if item['name'] == name:
                self.items.remove(item)
                return True
        return False
    def empty(self):
        return len(self.items) == 0
    def total(self):
        return sum(item['price'] for item in self.items)
    def show(self):
        if self.empty():
            print('Gio hang trong')
        else:
            for item in self.items:
                print(f"{item['name']} - {item['price']}")
    def clear(self):
        self.items.clear()

cart = Shopping()
while True:
    print("\n===== MENU =====")
    print("2. Them san pham")
    print("3. Xoa san pham theo ten")
    print("4. Kiem tra gio hang rong")
    print("5. Tinh tong tien")
    print("6. Hien thi gio hang")
    print("7. Xoa toan bo gio hang")

    choice = input("Nhap lua chon (2-7): ")
    if choice == '2':
        name = input("Nhap ten san pham: ")
        price = float(input("Nhap gia san pham: "))
        cart.add(name, price)
    elif choice == '3':
        name = input("Nhap ten san pham can xoa: ")
        if not cart.remove(name):
            print("Khong tim thay san pham")
    elif choice == '4':
        print('Gio hang rong' if cart.empty() else 'Gio hang khong rong')
    elif choice == '5':
        print('Tong tien', cart.total())
    elif choice == '6':
        cart.show()
    elif choice == '7':
        cart.clear()
        print('da xoa')
    else:
        print("Lua chon khong hop le")

    exit_choice = input("Ban co muon thoat? ")
    if exit_choice.lower() == "yes":
        break
                
```

```python
#W8A8
class Bank:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def withdraw(self,x):
        self.balance -= x
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
owner, balance = input().split()
balance = float(balance)
account = Bank(owner, balance)
n = int(input())
for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == 'DEPOSIT':
        account.deposit(x)
    elif cmd == 'WITHDRAW':
        account.withdraw(x)
print(f"{account.balance:.2f}")
```

```python
#W8A9
class Student:
    def __init__(self,name):
        self.name = name
        self.scores = []
    def add(self, subject, score):
        self.scores.append(score)
    def avg(self):
        return sum(self.scores)/len(self.scores)
    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        elif a >= 6.5:
            return "Good"
        elif a >= 5:
            return "Average"
        else:
            return "Poor"
name = input().strip()
n = int(input())
student = Student(name)
for _ in range(n):
    subject, score = input().split()
    student.add(subject, float(score))

avg_score = student.avg()
rank = student.rank()

print(f"{student.name} {avg_score:.2f} {rank}")
    
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == y)
n = int(input())
lib = Library()

for _ in range(n):
    line = input().strip()

    if line.startswith("ADD"):
        data = line[4:]          
        title, author, year = data.split(";")
        lib.add(Book(title, author, int(year)))

    elif line.startswith("COUNTYEAR"):
        year = int(line.split()[1])
        print(lib.find_by_year(year))

    elif line.startswith("COUNT"):
        author = line[6:]        
        print(lib.count_by_author(author))
```
